```python
###agent_simple_memory

import google.generativeai as genai
import os
from dotenv import load_dotenv

load_dotenv()

genai.configure(api_key="AIzaSyCd1d8Vn56e_Ul-KYUTdYEF5wdSl9EGtj4")

model = genai.GenerativeModel("gemini-2.5-flash-lite")
```

```python
#                      4           3                   2                1
memory = []
```

```python
def agent(query):
  memory.append("User :"+ query)
  context ="\n".join(memory[-3:])
  print("--------------------------------")
  print(context)
  print("--------------------------------")
  system = "You are an AI Agent. Remembers past conversation but keep answers short"
  prompt = system + "\n" + context
  response = model.generate_content(prompt)
  memory.append("Agent :" + response.text)
  print("--------------------------------")
  print(memory)
  print("--------------------------------")

  return response.text
```

```python
print("Type 'exit' to stop")
while True:
  q = input("User :")
  if q.lower() == "exit":
    break
  print("Agent :", agent(q))
```
